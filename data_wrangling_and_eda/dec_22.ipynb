```python
import pandas as pd
import math
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['PassengerId', 'Ticket', 'Cabin', 'Name'], axis=1, inplace=True)
```

```python
df
```

```python
## Use Pclass to fill in missing values in Age with the assumption that age and pclass are related

## For example, as you get older, you have more money to spend on first class travel compared to when you are younger.
```

```python
df.Pclass.corr(df.Age)
```

```python
fill_age_dict = df.groupby('Pclass')['Age'].median().to_dict()
```

```python
#key -> value
```

```python
fill_age_dict.get(3)
```

```python
def fill_age(row):
    if (math.isnan(row['Age'])):
        return fill_age_dict.get(row['Pclass'])
    return row['Age']
```

```python
df['Age'] = df.apply(fill_age, axis=1)
```

```python
df.info()
```

```python
df.Embarked.value_counts()
```

```python
df['Embarked'] = df.Embarked.fillna('S')
```

```python
df.info()
```

```python
df
```

```python
sex_mapped = pd.get_dummies(df['Sex'], drop_first=True)
```

```python
embarked_mapped = pd.get_dummies(df['Embarked'], drop_first=True)
```

```python
df = pd.concat([df, sex_mapped, embarked_mapped], axis=1).drop(['Sex', 'Embarked'], axis=1)
```
