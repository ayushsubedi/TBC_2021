I want the name list of titanic passengers sorted by their age (ascending)

```python
df.sort_values('Age', ascending=False)[['Age', 'Name']]
```

```python
df.sort_values(by = ['Fare', 'Name'])[['Fare', 'Name', 'Survived', 'Age']].head(20)
```

```python
df.sort_values('Age')
```

Gender wise average fare chai kati ho ta?

```python
df.groupby('Sex')[['Fare']].median()
```

```python
df.groupby('Sex')[['Survived']].sum()
```

# for age < 10 how what is the distribution of gender that survived?

```python
df[df.Age<10].groupby('Sex')[['Survived']].sum()
```
